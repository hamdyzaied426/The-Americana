#### Do students prefer universities with low tuition and fees ?

```python
question4 = cleaned_amrican[['Tuition and fees, 2013-14', 'Admissions yield - total']]
question4.head()
```

```python
sns.scatterplot(data=question4, x='Tuition and fees, 2013-14', y='Admissions yield - total')
```

```python
numOfenrolled_20000 = question4[(question4['Tuition and fees, 2013-14'] < 20000) & (question4['Admissions yield - total'] > 50)]
numOfenrolled_20000.describe()
```

```python
numOfenrolled_40000 = question4[(question4['Tuition and fees, 2013-14'] < 40000) & (question4['Tuition and fees, 2013-14'] >= 20000) & (question4['Admissions yield - total'] > 50)]
numOfenrolled_40000.describe()
```

```python
numOfenrolled_50000 = question4[(question4['Tuition and fees, 2013-14'] < 50000) & (question4['Tuition and fees, 2013-14'] >= 40000) & (question4['Admissions yield - total'] > 50)]
numOfenrolled_50000.describe()
```

```python
Tuition = {'less than 20000':len(numOfenrolled_20000),
           'less than 40000':len(numOfenrolled_40000),
           'less than 50000':len(numOfenrolled_50000)}

plt.pie(x=Tuition.values(), labels=Tuition.keys(), radius=1.2, shadow=True, autopct='%1.1f%%', textprops={'fontsize': 14, 'color':'w'})
plt.title("Tuition and fees with rate of admission > 0.5", color='w')
```

#### Do students prefer a university based on the possibility of a higher, additional academic degree in the same university?

```python
degree_offers = ['Offers Less than one year certificate',
                           'Offers One but less than two years certificate', 
                           "Offers Associate's degree",
                           "Offers Two but less than 4 years certificate",
                           'Offers Postbaccalaureate certificate',
                           "Offers Master's degree",
                           "Offers Post-master's certificate",
                           "Offers Doctor's degree - research/scholarship",
                           "Offers Doctor's degree - professional practice",
                           "Offers Doctor's degree - other",
                            'Offers Other degree']

question7 = cleaned_amrican[degree_offers + ['Admissions yield - total']]


question7.head()
```

```python
higher_degree = ["Offers Master's degree",
                "Offers Doctor's degree - research/scholarship",
                "Offers Doctor's degree - professional practice",
                "Offers Doctor's degree - other"]

for degree in higher_degree:
    sns.boxplot(data=question7, x=degree, y='Admissions yield - total')
    plt.show()
```

```python
other_degree = list(set(degree_offers) - set(higher_degree))
other_degree
```

```python
for degree in other_degree:
    sns.boxplot(data=question7, x=degree, y='Admissions yield - total')
    plt.show()
```
